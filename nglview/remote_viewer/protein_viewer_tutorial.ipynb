# Aim: Lightweight protein viewer for remote cluster

Need:

- Jupyter notebook
    ```bash
    conda install jupyter notebook
    ```
- pytraj (AmberTools): Used for loading pdb files or trajectories from disk

- nglview
    ```bash
    pip install git+https://github.com/arose/nglview
    ```
- a modern web browser (Chrome, Firefox ...)
